#### Q2.
Sobre os dados de copa do mundo, qual a distribuição de público presente nos jogos? Isso pode ser respondido com um histograma com os dados da coluna `comparecimento`.  

Lembre-se que alguns jogos estavam com público 0 incorretamente, que tal remover essas ocorrências para não atrapalhar sua visualzação?

Você deve implementar essa visualização nas três bibliotecas que vimos:
* Matplotlib
* Seaborn
* Plotly

Garanta que o gráfico tenha pelo menos os atributos de título e rótulos de dimensão.

````python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('wc_formatado.csv')
df_filtrado = df[df['comparecimento'] > 0]

plt.figure(figsize=(10, 6))
plt.hist(df_filtrado['comparecimento'], bins=30, color='steelblue', edgecolor='black')
plt.title('Distribuição do Público nos Jogos da Copa do Mundo')
plt.xlabel('Comparecimento (público)')
plt.ylabel('Frequência (número de jogos)')
plt.tight_layout()
plt.show()
````

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('wc_formatado.csv')
df_filtrado = df[df['comparecimento'] > 0]

plt.figure(figsize=(10, 6))
sns.histplot(data=df_filtrado, x='comparecimento', bins=30, color='mediumseagreen')
plt.title('Distribuição do Público nos Jogos da Copa do Mundo')
plt.xlabel('Comparecimento (público)')
plt.ylabel('Frequência (número de jogos)')
plt.tight_layout()
plt.show()
````

````python
import pandas as pd
import plotly.express as px

df = pd.read_csv('wc_formatado.csv')
df_filtrado = df[df['comparecimento'] > 0]

fig = px.histogram(
    df_filtrado,
    x='comparecimento',
    nbins=30,
    title='Distribuição do Público nos Jogos da Copa do Mundo',
    labels={'comparecimento': 'Comparecimento (público)', 'count': 'Frequência (número de jogos)'}
)
fig.update_layout(yaxis_title='Frequência (número de jogos)')
fig.show()
````
